```python
import pandas as pd
import numpy as np
data=pd.read_csv("/content/house-price-parquet.csv")
print(data.head())
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
#check duplicate
duplicate_rows=data[data.duplicated()]
print(duplicate_rows)
```

```python
#drop duplicate
data_cleaned=data.drop_duplicates()
```

```python
#remove outlayouts(z_score)
from scipy.stats import zscore
numeric_cols=data_cleaned.select_dtypes(include=[np.number]).columns
z_score=zscore(data[numeric_cols])
z_score_df=pd.DataFrame(z_score,columns=numeric_cols)
outliers = (np.abs(z_score_df)>3).any(axis=1)
data_cleaned = data[~outliers]
data_cleaned.reset_index(drop=True, inplace=True)

print(data_cleaned.head())
print("Original ahape:", data.shape)
print("Cleaned shape:", data_cleaned.shape)
```

```python
data.columns
```

```python
#data label encodinh
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in data_cleaned.select_dtypes(include='object').columns:
    data_cleaned.loc[:, col] = le.fit_transform(data_cleaned[col])
print(data_cleaned.head())
```

```python
x=data_cleaned.drop('price',axis=1)
y=data_cleaned['price']
print(x.head())
print(y.head)
```

```python
#feature selection
from sklearn.feature_selection import SelectKBest, mutual_info_regression

selector = SelectKBest(score_func=mutual_info_regression, k=10)  # select top 10 features
selector.fit(x, y)

selected_cols = x.columns[selector.get_support()]
x_selected = x[selected_cols]

print("Top selected features:\n", selected_cols)
x_selected.head()
```

```python
#skew_normalization
from sklearn.preprocessing import PowerTransformer
pt=PowerTransformer()
cols=['area', 'bedrooms', 'bathrooms', 'stories', 'mainroad', 'guestroom',
       'airconditioning', 'parking', 'prefarea', 'furnishingstatus']
x_selected.loc[:, cols] = pt.fit_transform(x_selected[cols])
skew_value = x_selected[cols].skew()
print(skew_value)
print(x_selected.head())
```

```python
#scalling
from sklearn.preprocessing import MinMaxScaler
import pandas as pd

scaler = MinMaxScaler()
x_scaled_array = scaler.fit_transform(x_selected)
x_scaled = pd.DataFrame(x_scaled_array, columns=x_selected.columns)
print(x_scaled.head())
```

```python
#join the x_scaled and y
new_data=pd.concat([x_scaled,y],axis=1)
print(new_data.head())
```

```python
#outliet_using_scaling
from scipy.stats import zscore
z_score=zscore(new_data)
zscore_df=pd.DataFrame(z_score,columns=new_data.columns)
outliers=(np.abs(zscore_df)>3).any(axis=1)
new_data=new_data[~outliers]
new_data.reset_index(drop=True,inplace=True)

print("new_data:",new_data.shape)
print(new_data.head())
```

```python
x_new=new_data.drop('price',axis=1)
y_new=new_data['price']
print(x_new.head())
print(y_new.head())
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_new,y_new,test_size=0.2)
print(x_train.head())
print(y_train.head())
```
